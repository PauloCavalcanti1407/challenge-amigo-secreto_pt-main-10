```python
# Importar bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# Configurações para visualizações
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Etapa 1: Extração (Extraction)
# Extrair dados de uma "API" no GitHub (URL direta para CSV)
url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
print("Dados extraídos com sucesso!")
print(f"Shape inicial: {df.shape}")
print(df.head())

# Etapa 2: Transformação (Transformation)
# Limpeza e preparação dos dados
# Verificar valores ausentes
print("\nValores ausentes por coluna:")
print(df.isnull().sum())

# Tratar coluna 'TotalCharges' (alguns valores são strings vazias, convertê-los para NaN e depois imputar)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

# Remover coluna 'customerID' (não necessária para análise)
df.drop('customerID', axis=1, inplace=True)

# Codificar variáveis categóricas para análise (usando LabelEncoder para simplicidade)
le = LabelEncoder()
categorical_cols = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService',
                    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
                    'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn']
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Verificar shape após transformação
print(f"\nShape após transformação: {df.shape}")
print(df.head())

# Etapa 3: Carga (Load) - Aqui, os dados estão prontos para análise; em um cenário real, carregaria em um banco de dados
# Para este desafio, prosseguimos para EDA

# Etapa 4: Análise Exploratória de Dados (EDA)
# Estatísticas descritivas
print("\nEstatísticas descritivas:")
print(df.describe())

# Distribuição da variável alvo (Churn)
plt.figure()
sns.countplot(x='Churn', data=df)
plt.title('Distribuição de Evasão de Clientes')
plt.show()

# Correlação entre variáveis
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Mapa de Correlação')
plt.show()

# Análise de churn por contrato
plt.figure()
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Evasão por Tipo de Contrato')
plt.show()

# Análise de churn por serviço de internet
plt.figure()
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title('Evasão por Serviço de Internet')
plt.show()

# Boxplot de Tenure vs Churn
plt.figure()
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title('Tempo de Permanência vs Evasão')
plt.show()

# Insights iniciais (serão detalhados no relatório)
print("\nInsights preliminares da EDA:")
print("- Clientes com contratos mensais têm maior taxa de churn.")
print("- Serviços de internet DSL e fibra óptica mostram diferenças em churn.")
print("- Clientes com tenure menor tendem a evadir mais.")
```
